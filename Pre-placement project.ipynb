### Level 1 Analysis - Separate the Categorical and Numerical Column

```python
numeric_columns = []
categorical_columns = []

for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        numeric_columns.append(column)
    else:
        categorical_columns.append(column)

table = [numeric_columns, categorical_columns]
print(tabulate({"numerical":numeric_columns,
                "categorical": categorical_columns}, headers = ["numerical", "categorical"]))
```

### Performing Uni - Variant Analysis for Categorical column

```python
def info_of_cat(col):
    print(f"Unique values in {col} are :{df[col].unique()}")
    print(f"Mode of {col} is : {df[col].mode()[0]}")
    print(f"Number of missing values in {col} is : {df[col].isnull().sum()}")
```

##### User-defined function for level 1 analysis of categorical columns
This function will check for the following 
- Unique values

#### To find the relationship between one numeric and one categorical column 
- We can use Bar plot , to find the relatioship betweeen on numeric and one categorical column

#### Bivariant analysis between City wise total sales

```python
City_Sales = df.groupby(by=['City'])['Total_Sales'].sum()
pd.DataFrame(City_Sales)
City_Sales.reset_index()
```

```python
plt.figure(figsize=(15, 9))
City_Sales.plot(kind='bar', color='skyblue')

plt.xlabel('State')
plt.ylabel('Total number of sales')
plt.title('Number of Sales in Different Cities')
plt.show()
```

Here the State of New York and charleston has the high sales compared to other cities

#### Bivariant Analysis between Retailer and Total Sales

```python
df.groupby(['Retailer'])['Total_Sales'].aggregate(lambda x: x.sum()).plot(kind='bar' ,xlabel='Retailers',ylabel='Total sales taken part',
                                      title='Retailer wise total sales ',figsize=(16,9),color=['lightgreen','cyan','gray','orange','pink','brown'])
```

Here the Foot Locker and West Gear has the Maximum number of sales 

#### Bivariant Analysis between Region and Total Sales

```python
df.groupby(['Region'])['Total_Sales'].aggregate(lambda x: x.sum()).plot(kind='bar' ,xlabel='Region',ylabel='Total sales taken part',
                                      title='Region wise total sales ',figsize=(16,9),color=['lightgreen','cyan','gray','orange','pink'])
```

Here we can see that the west region has the highest sales compared to other regions.

#### Bivariant Analysis of Sales method wise Total sales

```python
df.groupby(['Sales_Method'])['Total_Sales'].aggregate(lambda x: x.sum()).plot(kind='bar' ,xlabel='Sales Method',ylabel='Total sales taken part',
                                      title= 'Sales method wise total sales ',figsize=(16,9),color=['lightgreen','cyan','gray'])
```

in store sales method shows the highest sales than compared to online and outlet
